```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
xls=pd.ExcelFile('millenniumofdata_v3_final.xlsx')
df1 = pd.read_excel(xls,'A2. Pop of Eng & GB 1086-1870',skiprows=7)
df2 = pd.read_excel(xls,'A3. Eng. Agriculture 1270-1870',skiprows=10)
```

```python
df2.head()
```

```python
h=[i[:-2] for i in df2.columns if i[-2:]=='.1']
h
```

```python
df2.loc[0,'Livestock Output %']='current prices'
df2.loc[0,'Livestock Output %.1']='constant 1700 prices'
#(df2[h+[i+'.1' for i in h]].columns +' '+ df2[h+[i+'.1' for i in h]].iloc[0]).str.replace('.1','').to_dict()
df2 = df2[[i for i in df2.columns if not 'Unnamed' in i]].rename(columns=(df2[h+[i+'.1' for i in h]].columns +' '+ df2[h+[i+'.1' for i in h]].iloc[0]).str.replace('.1','').to_dict()).drop(0)
```

```python
df2 = df2.astype(float)
df2.Year=df2.Year.astype(int)
df2.head()
```

```python
sns.lineplot(data=df1,x='Year',y='Population of England, millions')
sns.lineplot(data=df1,x='Year',y='Population of Great Britain, millions')
#df1.head()
```

```python
#sns.lineplot(data=df2, x='Year',y='Total Arable and Sown Acreage')
#sns.lineplot(data=df2, x='Year',y='Wheat million bushels')
df2['quarters']=(df2['Year']//25)*25
df2_mean_cols=[i for i in df2.columns.tolist() if ('bushels per acre' in i or 'prices' in i or 'Year' in i) and not (i=='Potatoes Crop Yield (bushels per acre)')]
#print(df2_mean_cols)
df2_sum_cols=[i for i in df2.columns if not i in df2_mean_cols]
#df3=df2.groupby('quarters',as_index=False).agg({'Year':'mean','Total Arable and Sown Acreage':'mean',
#                                        'Wheat Crop Yield (bushels per acre)':'mean', 'Wheat million bushels':'sum'})

df3=df2.groupby('quarters',as_index=False).agg({i:'mean' for i in df2_mean_cols} | {i:'sum' for i in df2_sum_cols})
df2_mean_cols = [i for i in df2_mean_cols if not i=='Year' and not i=='quarters']
df3.Year=df3.Year.astype(int)
df3.drop('quarters',axis=1)
df3.columns
```

```python
df4 = df3[df2_mean_cols + ['Year']].melt(id_vars='Year', value_vars=df2_mean_cols)
sns.lineplot(data=df4,x='Year',y='value',hue='variable')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
#plt.xticks(rotation=45);
```
